# 🌾 AgriFlow AI

## Intelligent Cascaded Machine Learning Pipeline for Precision Agriculture

### Project Objective

AgriFlow AI is an intelligent decision support system designed to assist farmers in making informed agricultural decisions using a cascaded machine learning pipeline.

The system performs three sequential tasks:

- 🌱 Crop Recommendation
- 🌾 Yield Prediction
- 🌿 Fertilizer Recommendation

Unlike traditional machine learning projects that solve a single task, AgriFlow AI integrates multiple machine learning models where the output of one stage becomes the input to the next, enabling end-to-end agricultural decision support.
