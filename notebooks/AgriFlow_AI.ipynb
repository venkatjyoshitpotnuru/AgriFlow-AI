# 🌾 AgriFlow AI Pipeline Design

## Problem Statement

Modern agriculture involves multiple interconnected decisions rather than a single prediction. Farmers need to determine the most suitable crop based on environmental conditions, estimate the expected yield of that crop, and finally decide the most appropriate fertilizer to maximize productivity.

A single machine learning model cannot effectively solve all these dependent tasks. Therefore, AgriFlow AI adopts a **Cascaded Machine Learning Pipeline**, where each model performs a specialized task and passes its output to the next stage.

---

## Pipeline Overview

The AgriFlow AI pipeline consists of three sequential machine learning models:

### 🌱 Stage 1 – Crop Recommendation

**Objective:**

Recommend the most suitable crop based on soil nutrients and environmental conditions.

**Input Features**

- Nitrogen (N)
- Phosphorus (P)
- Potassium (K)
- Temperature
- Humidity
- Soil pH
- Rainfall

**Output**

Recommended Crop

---

### 🌾 Stage 2 – Yield Prediction

**Objective**

Predict the expected crop yield using the crop recommended in Stage 1 along with agricultural and climatic information.

**Input Features**

- Recommended Crop
- Crop Year
- Season
- State
- Area
- Annual Rainfall
- Fertilizer Usage
- Pesticide Usage
- Average Temperature
- Maximum Temperature
- Minimum Temperature

**Output**

Expected Crop Yield

---

### 🌿 Stage 3 – Fertilizer Recommendation

**Objective**

Recommend the most suitable fertilizer using the predicted crop yield together with soil characteristics and crop information.

**Input Features**

- Recommended Crop
- Predicted Yield
- Soil Type
- Soil Moisture
- Soil pH
- Nitrogen Level
- Phosphorus Level
- Potassium Level
- Temperature
- Humidity
- Rainfall
- Crop Growth Stage
- Irrigation Type
- Previous Crop

**Output**

Recommended Fertilizer

---

## Final Output

The final system generates a farmer recommendation report containing:

- Recommended Crop
- Expected Yield
- Recommended Fertilizer

This enables farmers to make informed agricultural decisions through an integrated AI-based decision support system.

# System Architecture

```mermaid
flowchart TD

A[Farmer Input Data]

A --> B[Stage 1<br/>Crop Recommendation]

B --> C[Recommended Crop]

C --> D[Stage 2<br/>Yield Prediction]

D --> E[Expected Yield]

E --> F[Stage 3<br/>Fertilizer Recommendation]

F --> G[Farmer Recommendation Report]
```

## Why Cascaded Machine Learning?

Traditional machine learning projects solve only one prediction task.

However, agricultural decision-making involves multiple dependent stages.

- Crop recommendation must be performed before estimating crop yield.
- Yield estimation provides additional information for fertilizer recommendation.
- Each stage improves the quality of the next stage by providing context-specific predictions.

This modular design improves interpretability, maintainability, and reflects how real-world agricultural decision support systems operate.
