```python
"""
Cascaded ML Pipeline: Credit Risk Decision Engine
==================================================
Team project - End-to-end pipeline where three models work in sequence.

STAGE 1 (Classification) : Probability of Default (PD)
STAGE 2 (Regression)     : Loss Given Default (LGD) severity estimate
STAGE 3 (Classification) : Final lending decision (Approve / Review / Reject)
                           using PD + LGD + Exposure as engineered inputs

Dataset: Statlog (German Credit Data), UCI Machine Learning Repository
Mirror used (public, direct CSV): https://raw.githubusercontent.com/jbrownlee/Datasets/master/german.csv
Original source: https://archive.ics.uci.edu/dataset/144/statlog+german+credit+data
1000 loan applicants, 20 attributes, target = credit risk (good/bad)
"""
```

```python
import pandas as pd
import numpy as np
import json
import warnings
warnings.filterwarnings("ignore")
```

```python
import matplotlib
matplotlib.use("Agg")
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier, GradientBoostingRegressor
from sklearn.metrics import (accuracy_score, precision_score, recall_score, f1_score,
                              roc_auc_score, confusion_matrix, classification_report,
                              mean_absolute_error, mean_squared_error, r2_score)
from xgboost import XGBClassifier
```

```python
RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)
```

```python
OUT = "outputs/"
metrics_summary = {}
```

```python
# ============================================================
# STAGE 0: LOAD + CLEAN DATA
# ============================================================
COLUMNS = [
    "checking_status", "duration", "credit_history", "purpose", "credit_amount",
    "savings_status", "employment", "installment_commitment", "personal_status",
    "other_parties", "residence_since", "property_magnitude", "age",
    "other_payment_plans", "housing", "existing_credits", "job",
    "num_dependents", "own_telephone", "foreign_worker", "class"
]
```

```python
df = pd.read_csv("german.csv", header=None, names=COLUMNS)
df["default"] = (df["class"] == 2).astype(int)   # 1 = bad credit / default
df = df.drop(columns=["class"])
```

```python
print("="*70)
print("STAGE 0: DATA LOADED")
print("="*70)
print(f"Shape: {df.shape}")
print(f"Default rate: {df['default'].mean():.3f}")
assert df.isnull().sum().sum() == 0, "unexpected nulls"
metrics_summary["dataset"] = {"n_rows": int(df.shape[0]), "n_features": int(df.shape[1]-1),
                               "default_rate": round(float(df["default"].mean()), 4)}
```

```python
# ============================================================
# STAGE 0b: EDA (a few key plots saved for the report)
# ============================================================
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
sns.histplot(df["credit_amount"], bins=30, ax=axes[0], color="#8B4513")
axes[0].set_title("Credit Amount Distribution")
sns.histplot(df["age"], bins=20, ax=axes[1], color="#2F4F4F")
axes[1].set_title("Applicant Age Distribution")
sns.barplot(x=df["default"].value_counts().index, y=df["default"].value_counts().values,
            ax=axes[2], color="#B22222")
axes[2].set_xticklabels(["Good (0)", "Default (1)"])
axes[2].set_title("Class Balance")
plt.tight_layout()
plt.savefig(OUT + "eda_overview.png", dpi=130)
plt.close()
```

```python
corr_features = ["duration", "credit_amount", "installment_commitment", "residence_since",
                  "age", "existing_credits", "num_dependents", "default"]
plt.figure(figsize=(7, 6))
sns.heatmap(df[corr_features].corr(), annot=True, fmt=".2f", cmap="RdBu_r", center=0)
plt.title("Correlation Matrix (numeric features)")
plt.tight_layout()
plt.savefig(OUT + "correlation_matrix.png", dpi=130)
plt.close()
```

```python
# ============================================================
# TRAIN / TEST SPLIT
# ============================================================
X = df.drop(columns=["default"])
y = df["default"]
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=RANDOM_STATE, stratify=y
)
```

```python
cat_cols = X.select_dtypes(include="object").columns.tolist()
num_cols = X.select_dtypes(exclude="object").columns.tolist()
```

```python
def make_preprocessor():
    return ColumnTransformer([
        ("num", StandardScaler(), num_cols),
        ("cat", OneHotEncoder(handle_unknown="ignore"), cat_cols)
    ])
```

```python
# ============================================================
# STAGE 1: CLASSIFICATION -- Probability of Default (PD)
# ============================================================
print("\n" + "="*70)
print("STAGE 1: PD CLASSIFICATION MODEL")
print("="*70)
```

```python
model1 = Pipeline([
    ("prep", make_preprocessor()),
    ("clf", XGBClassifier(
        n_estimators=250, max_depth=3, learning_rate=0.05,
        subsample=0.8, colsample_bytree=0.8, eval_metric="logloss",
        random_state=RANDOM_STATE, scale_pos_weight=(y_train==0).sum()/(y_train==1).sum()
    ))
])
model1.fit(X_train, y_train)
```

```python
pd_pred_proba_test = model1.predict_proba(X_test)[:, 1]
pd_pred_class_test = (pd_pred_proba_test >= 0.5).astype(int)
```

```python
m1_metrics = {
    "accuracy": round(accuracy_score(y_test, pd_pred_class_test), 4),
    "precision": round(precision_score(y_test, pd_pred_class_test), 4),
    "recall": round(recall_score(y_test, pd_pred_class_test), 4),
    "f1": round(f1_score(y_test, pd_pred_class_test), 4),
    "roc_auc": round(roc_auc_score(y_test, pd_pred_proba_test), 4),
}
print(json.dumps(m1_metrics, indent=2))
metrics_summary["model1_PD_classification"] = m1_metrics
```

```python
cm1 = confusion_matrix(y_test, pd_pred_class_test)
plt.figure(figsize=(4.5, 4))
sns.heatmap(cm1, annot=True, fmt="d", cmap="Oranges",
            xticklabels=["Good", "Default"], yticklabels=["Good", "Default"])
plt.title("Model 1 (PD) - Confusion Matrix")
plt.ylabel("Actual"); plt.xlabel("Predicted")
plt.tight_layout()
plt.savefig(OUT + "model1_confusion_matrix.png", dpi=130)
plt.close()
```

```python
# ============================================================
# STAGE 2: REGRESSION -- Loss Given Default (LGD) severity
# ============================================================
print("\n" + "="*70)
print("STAGE 2: LGD REGRESSION MODEL")
print("="*70)
```

```python
def base_lgd(row):
    coll_map = {
        "A121": 0.25,  # real estate
        "A122": 0.38,  # building society savings / life insurance
        "A123": 0.48,  # car or other
        "A124": 0.68,  # unknown / no property
    }
    lgd = coll_map.get(row["property_magnitude"], 0.5)
    if row["housing"] == "A152":
        lgd -= 0.05
    elif row["housing"] == "A151":
        lgd += 0.04
    elif row["housing"] == "A153":
        lgd += 0.02
    emp_map = {"A71": 0.05, "A72": 0.03, "A73": 0.0, "A74": -0.03, "A75": -0.05}
    lgd += emp_map.get(row["employment"], 0.0)
    lgd += 0.015 * (row["existing_credits"] - 1)
    return lgd
```

```python
df["lgd_base"] = df.apply(base_lgd, axis=1)
noise = np.random.normal(0, 0.06, size=len(df))
df["lgd_severity"] = np.clip(df["lgd_base"] + noise, 0.05, 0.95)
```

```python
defaulted = df[df["default"] == 1].copy()
Xd = defaulted.drop(columns=["default", "lgd_base", "lgd_severity"])
yd = defaulted["lgd_severity"]
```

```python
Xd_train, Xd_test, yd_train, yd_test = train_test_split(
    Xd, yd, test_size=0.25, random_state=RANDOM_STATE
)
```

```python
model2 = Pipeline([
    ("prep", make_preprocessor()),
    ("reg", GradientBoostingRegressor(
        n_estimators=200, max_depth=3, learning_rate=0.05, random_state=RANDOM_STATE
    ))
])
model2.fit(Xd_train, yd_train)
lgd_pred_test = model2.predict(Xd_test)
```

```python
m2_metrics = {
    "MAE": round(mean_absolute_error(yd_test, lgd_pred_test), 4),
    "RMSE": round(mean_squared_error(yd_test, lgd_pred_test) ** 0.5, 4),
    "R2": round(r2_score(yd_test, lgd_pred_test), 4),
    "n_train_defaults": int(len(Xd_train)),
    "n_test_defaults": int(len(Xd_test)),
}
print(json.dumps(m2_metrics, indent=2))
metrics_summary["model2_LGD_regression"] = m2_metrics
```

```python
plt.figure(figsize=(5, 5))
plt.scatter(yd_test, lgd_pred_test, alpha=0.6, color="#4682B4")
plt.plot([0, 1], [0, 1], "r--")
plt.xlabel("Actual LGD severity"); plt.ylabel("Predicted LGD severity")
plt.title("Model 2 (LGD) - Actual vs Predicted")
plt.tight_layout()
plt.savefig(OUT + "model2_actual_vs_pred.png", dpi=130)
plt.close()
```

```python
# ============================================================
# STAGE 3: CLASSIFICATION -- Final lending decision
# ============================================================
print("\n" + "="*70)
print("STAGE 3: DECISION CLASSIFICATION MODEL")
print("="*70)
```

```python
df["pd_hat"] = model1.predict_proba(X)[:, 1]
df["lgd_hat"] = model2.predict(X)
df["lgd_hat"] = df["lgd_hat"].clip(0.05, 0.95)
```

```python
df["exposure"] = df["credit_amount"]
df["expected_loss"] = df["pd_hat"] * df["lgd_hat"] * df["exposure"]
```

```python
low_thr = df["expected_loss"].quantile(0.55)
high_thr = df["expected_loss"].quantile(0.85)
```

```python
def policy_label(el):
    if el < low_thr:
        return 0
    elif el < high_thr:
        return 1
    else:
        return 2
```

```python
df["decision_label"] = df["expected_loss"].apply(policy_label)
```

```python
flip_mask = np.random.rand(len(df)) < 0.07
flip_shift = np.random.choice([-1, 1], size=len(df))
df.loc[flip_mask, "decision_label"] = (df.loc[flip_mask, "decision_label"] + flip_shift[flip_mask]).clip(0, 2)
```

```python
decision_features = ["pd_hat", "lgd_hat", "expected_loss", "exposure", "duration", "age",
                      "installment_commitment", "existing_credits"]
X3 = df[decision_features]
y3 = df["decision_label"]
```

```python
X3_train, X3_test, y3_train, y3_test = train_test_split(
    X3, y3, test_size=0.25, random_state=RANDOM_STATE, stratify=y3
)
```

```python
model3 = RandomForestClassifier(
    n_estimators=300, max_depth=6, random_state=RANDOM_STATE, class_weight="balanced"
)
model3.fit(X3_train, y3_train)
y3_pred = model3.predict(X3_test)
```

```python
m3_metrics = {
    "accuracy": round(accuracy_score(y3_test, y3_pred), 4),
    "macro_f1": round(f1_score(y3_test, y3_pred, average="macro"), 4),
}
print(json.dumps(m3_metrics, indent=2))
print(classification_report(y3_test, y3_pred, target_names=["Approve", "Review", "Reject"]))
metrics_summary["model3_decision_classification"] = m3_metrics
metrics_summary["model3_classification_report"] = classification_report(
    y3_test, y3_pred, target_names=["Approve", "Review", "Reject"], output_dict=True
)
```

```python
cm3 = confusion_matrix(y3_test, y3_pred)
plt.figure(figsize=(5, 4.5))
sns.heatmap(cm3, annot=True, fmt="d", cmap="Purples",
            xticklabels=["Approve", "Review", "Reject"], yticklabels=["Approve", "Review", "Reject"])
plt.title("Model 3 (Decision) - Confusion Matrix")
plt.ylabel("Actual"); plt.xlabel("Predicted")
plt.tight_layout()
plt.savefig(OUT + "model3_confusion_matrix.png", dpi=130)
plt.close()
```

```python
imp = pd.Series(model3.feature_importances_, index=decision_features).sort_values()
plt.figure(figsize=(6, 4))
imp.plot(kind="barh", color="#556B2F")
plt.title("Model 3 - Feature Importance")
plt.tight_layout()
plt.savefig(OUT + "model3_feature_importance.png", dpi=130)
plt.close()
```

```python
# ============================================================
# OVERALL SYSTEM EVALUATION: cascade vs. naive single-model policy
# ============================================================
print("\n" + "="*70)
print("OVERALL SYSTEM EVALUATION")
print("="*70)
```

```python
test_idx = X3_test.index
naive_reject = (df.loc[test_idx, "pd_hat"] >= 0.5).astype(int) * 2
```

```python
actual_default = df.loc[test_idx, "default"]
actual_exposure = df.loc[test_idx, "exposure"]
actual_lgd_realized = df.loc[test_idx, "lgd_severity"]
```

```python
def realized_loss(decisions):
    approved = decisions == 0
    loss = (approved & (actual_default == 1)) * actual_exposure * actual_lgd_realized
    return loss.sum()
```

```python
cascade_decisions = pd.Series(y3_pred, index=test_idx)
cascade_loss = realized_loss(cascade_decisions)
naive_loss = realized_loss(pd.Series(naive_reject.values, index=test_idx))
```

```python
cascade_approved_good_loans = int(((cascade_decisions == 0) & (actual_default == 0)).sum())
naive_approved_good_loans = int(((naive_reject == 0) & (actual_default == 0)).sum())
```

```python
system_eval = {
    "test_set_size": int(len(test_idx)),
    "cascade_realized_loss_on_test_set": round(float(cascade_loss), 2),
    "naive_PD_only_realized_loss_on_test_set": round(float(naive_loss), 2),
    "loss_reduction_pct": round(float((naive_loss - cascade_loss) / naive_loss * 100), 2) if naive_loss > 0 else None,
    "cascade_good_loans_approved": cascade_approved_good_loans,
    "naive_good_loans_approved": naive_approved_good_loans,
}
print(json.dumps(system_eval, indent=2))
metrics_summary["overall_system_evaluation"] = system_eval
```

```python
with open(OUT + "metrics_summary.json", "w") as f:
    json.dump(metrics_summary, f, indent=2)
```

```python
df.to_csv(OUT + "scored_dataset.csv", index=False)
print("\nAll metrics saved to outputs/metrics_summary.json")
print("Pipeline complete.")
```
